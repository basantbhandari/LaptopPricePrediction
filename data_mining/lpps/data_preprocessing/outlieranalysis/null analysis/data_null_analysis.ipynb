```python
#importing necessary libraries and uploading csv file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import io
from google.colab import files
uploaded = files.upload()
df_nullfile = pd.read_csv(io.BytesIO(uploaded['categorized_data.csv']))
df_nullfile
```

```python
#seeing no.of non-null values and datatype  in each column
df_nullfile.info()
```

```python
#checking number of null(NaN) values in each column
print("\nTotal missing value in each column in Dataframe:\n",df_nullfile.isnull().sum())
```

```python
#setting brandname as index(ie brandname instead of index(symbol) number)
df_nullfile.set_index("brandname",inplace=True)
```

```python
df_nullfile
```

```python
#we have total 10 columns now(excluding brandname)
#keeping only those rows which have alteast 5 valid values i.e. deleting rows with 5 or more inavlid(null) values
df_nullfile1=df_nullfile.dropna(thresh=5)
```

```python
df_nullfile1
```

```python
# replacing null value for ssd and hdd by 0 as when one has valid value other has 0 in most cases.
df_nullfile2=df_nullfile.fillna({'ssd':0,
                                'rom':0
})
```

```python
df_nullfile2
```

```python
#we have total 10 columns now(excluding brandname)
#keeping only those rows which have alteast 5 valid values i.e. deleting rows with 5 or more inavlid(null) values after replacing missing(null) values in rom and ssd with 0.
df_nullfile3=df_nullfile2.dropna(thresh=5)
```

```python
df_nullfile3
```

```python
#replacing null values of gpu by intel
df_nullfile4=df_nullfile3.fillna({'gpu':"Intel"
})
```

```python
df_nullfile4
```

```python
#checking number of null(NaN) values in each column after filling missing(null) values of gpu,rom,ssd
print("\nTotal missing value in each column in Dataframe:\n",df_nullfile4.isnull().sum())
```

As we can see now the total null value in gpu ,rom,ssd is 0.

```python
#reseting index to default
df_nullfile4.reset_index()
```

```python
#checking number of null(NaN) values in each column
print("\nTotal missing value in each column in Dataframe:\n",df_nullfile4.isnull().sum())
```

```python
# Impute display quality and cpucore with mode (Categorical Variable) by "Generalized imputation"
df_nullfile4['displayquality'].fillna(df_nullfile4['displayquality'].mode()[0], inplace = True)
df_nullfile4['cpucore'].fillna(df_nullfile4['cpucore'].mode()[0], inplace = True)
```

```python
# Impute ram with median(Numerical value) by "Generalized imputation"
df_nullfile4['ram'].fillna(df_nullfile4['ram'].median(), inplace = True)
```

```python
# Impute display with median(Numerical value) correlating it with brandname by "Special case imputation"
df_nullfile4['display'].fillna(df_nullfile4.groupby(["brandname"])['display'].transform('median'),inplace=True)
```

```python
# Impute display with mode(Numerical value) correlating it with both display and displayquality by "Special case imputation"
df_nullfile4['refreshrate'].fillna(df_nullfile4.groupby(["display","displayquality"])['refreshrate'].transform('median'),inplace=True)
```

```python
#checking number of null(NaN) values in each column
print("\nTotal missing value in each column in Dataframe:\n",df_nullfile4.isnull().sum())
```

As we can see even though we imputed refreshrate ,it still has null value.It is because we imputed  refreshrate using median correlating with pair of display and displayquality.For unique pair of display and display quality no median value was produced,so was still null.


```python
# Impute display with mode(Numerical value) correlating it with  displayquality only by "Special case imputation"
df_nullfile4['refreshrate'].fillna(df_nullfile4.groupby(["displayquality"])['refreshrate'].transform('median'),inplace=True)
```

```python
#checking number of null(NaN) values in each column
print("\nTotal missing value in each column in Dataframe:\n",df_nullfile4.isnull().sum())
```

Still,refreshrate has null value.It is because we imputed refreshrate using median correlating with  displayquality.For unique (100 non repeating ) displayquality no median value was produced,so was still null.

```python
# Impute display with mode(Numerical value) correlating it with  display only by "Special case imputation"
df_nullfile4['refreshrate'].fillna(df_nullfile4.groupby(["display"])['refreshrate'].transform('median'),inplace=True)
```

```python
#checking number of null(NaN) values in each column
print("\nTotal missing value in each column in Dataframe:\n",df_nullfile4.isnull().sum())
```

No any column has missing(null) value now.
