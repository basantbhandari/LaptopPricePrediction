```python
import requests
from bs4 import BeautifulSoup
```

```python

laptop_name=[]
laptop_price=[]
url="https://www.btplanet.com.np/"
req=requests.get(url)
content=BeautifulSoup(req.content,'html.parser')
name=content.find_all('h6',{"class":"card-title product-title"})
price=content.find_all('span',{"class":"discount-price"})
print("Laptops in page")
   
print(len(name))
print(len(price))
    
for i in name:
    laptop_name.append(i.text)
for i in price:
    laptop_price.append(i.text)
```

```python
for i in laptop_name:
    print(i)
```

```python
for i in laptop_price:
    print(i)
```

```python
for i in range(len(laptop_name)):
    print(laptop_name[i] + '\t ' + laptop_price[i])
    
    
    
```
